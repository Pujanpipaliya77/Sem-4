### 09) WAP to check whether two given strings are anagram or not using set.

```python
s1 = "hello"
s2 = "hi"

anagram = set(s1) == set(s2) and len(s1) == len(s2)

if anagram:
    print("words are anagram");
else:
    print("words are not anagram");
```
